# Task 1: Python Basics & Operators

```python
# User input:accept two integers from the user
a = int(input("Enter a number:"))
b = int(input("Enter a number:"))

# basic arithmetic operations (add, subtract, multiply, divide, modulus, power)
add = a + b
sub = a - b
mul = a * b
div = a / b
modulo = a % b
power = a ** b

# logical operators (>, <, ==, !=, and, or, not)
greater_than = (a > b)
smaller_than = (a < b)
equal = (a == b)
not_equal = (a != b)
logical_and = (a > 5) and (b < 5)
logical_or = (a < 7) or (b > 10)
logical_not = not(a == 100)

# printing result using formatted string
print(f"\n Arithmetic Operations")
print(f"Addition:{a} + {b} = {add}")
print(f"Subtraction:{a} - {b} = {sub}")
print(f"Multiplication:{a} * {b} = {mul}")
print(f"Division:{a} / {b} = {div}")
print(f"Modulo Division:{a} % {b} = {modulo}")
print(f"Power:{a} ** {b} = {power}")

print(f"\n Operators")
print(f"{a} > {b}: {greater_than}")
print(f"{a} < {b}: {smaller_than}")
print(f"{a} == {b}: {equal}")
print(f"{a} != {b}: {not_equal}")

print(f"\n Logical Operators")
print(f"AND: {logical_and}")
print(f"OR: {logical_or}")
print(f"Not: {logical_not}")
```

# Task 2: Lists and Arrays

```python
# list of 10 random integers
import random
import numpy as np
num = [random.randint(1, 100) for i in range(10)]
print(f"list:{num}")

num.append(4)  # Adding Number to a list
print(f"Add a element in list:{num}")
num.remove(4) # Removing Number from list
print(f"Removing element from list:{num}")

Max =max(num)  # Max Number of a array
print(f"Maximum Number:{Max}")
Min = min(num) # Min Number of a array
print(f"Minimun nUmber:{Min}")

sort = sorted(num) # Sorting array
print(f"Sorting an array:{sort}")

# Converting List to Numpy array
num_array = np.array(num)
Mean = np.mean(num_array)   # Mean of a array
Median= np.median(num_array) # Mediain of a array
sd = np.std(num_array) # Standard deviation of a array
print(f"Mean of a array:{Mean}")
print(f"Median of a array:{Median}")
print(f"Standard Deviation of a array:{sd}")
```

# Task 3: Dictionaries and Sets

```python
# Create a dictinary
student = {'name' : 'Snr', 'course' : 'Gen AI', 'marks' : 95}

# Add grade based on marks
marks = student['marks']
if marks >= 90:
    grade = 'A'
elif marks >= 60:
    grade = 'B'
else:
    grade = 'C'

student['grade'] = grade

# print all student data i.e. keys and values
for key, value in student.items():
    print(f"{key}: {value}")

# Create two sets of AI tools
set1 = {'ChatGPT', 'Claude', 'Gemini', 'Copilot'}
set2 = {'ChatGPT', 'DALL-E', 'Preplexity','Midjourney' }

print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
```

# Task 4: File Handling

```python
# Create and write details of students to ai_students.txt file
with open("ai_students.txt", "w") as file:
    file.write("Ravi,82,A\n")
    file.write("Anita,74,B\n")
    file.write("Suresh,91,A+\n")
    file.write("Priya,68,C\n")
    file.write("Kiran,88,A\n")

# Read the file and display students details with marks > 75
print("Students who scored above 75 marks:\n")

with open("ai_students.txt", "r") as file:
    for line in file:
        name, marks, grade = line.strip().split(",")
        if int(marks) > 75:
            print(f"Name: {name}, Marks: {marks}, Grade: {grade}")
```
